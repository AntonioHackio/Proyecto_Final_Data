# Resumen de Transformaciones Aplicadas a los Datos

## 1. TRANSFORMACIONES EN LEADS

### Limpieza y Normalización
- **Columnas**: Convertidas a minúsculas
- **Teléfonos**: 
  - Eliminado ".0" al final
  - Validación de longitud (máximo 9 dígitos)
- **Edad**: 
  - Conversión a tipo Int64 
  
- **Presupuesto**: 
  - Conversión de formato europeo (1.234,56 → 1234.56)
  - Redondeo a 2 decimales
- **Email**: Convertido a minúsculas
- **Campos de texto**: 
  - Aplicado `strip()` para eliminar espacios
  - `title()` en nombres, apellidos, ciudades, provincias

### Validaciones
- **Urgencia formación**: Limitada a escala 1-5
- **Lead score**: Limitado a escala 0-100
- **Duplicados**: Eliminados por email

### Nuevas Categorías Creadas
- **grupo_edad**: 
  - 18-25
  - 26-35
  - 36-45
  - 46-55
  - 56+
- **categoria_presupuesto**: 
  - Bajo (<3000)
  - Medio (3000-5000)
  - Alto (5000-7000)
  - Pro (>7000)
- **calidad_lead**: 
  - Bajo (0-40)
  - Medio (40-60)
  - Alto (60-80)
  - Excelente (80-100)

## 2. TRANSFORMACIONES EN CURSOS

### Limpieza de Datos
- Eliminadas filas sin Curso ID o completamente vacías
- Normalización de espacios en campos de texto
- Conversión de columnas numéricas:
  - Duración meses
  - Horas totales
  - Precios (base y descuento)
  - Plazas disponibles
  - Rating
  - Empleabilidad

### Procesamiento de Fechas y Booleanos
- Fechas formateadas a `YYYY-MM-DD`
- Certificación oficial: "Sí" → booleano `True/False`

### Métricas Derivadas Calculadas
- **Descuento_porcentaje**: `((precio_base - precio_descuento) / precio_base) * 100`
- **Horas_por_mes**: `horas_totales / duracion_meses`
- **Categoria_precio**: 
  - Económico (<2000€)
  - Medio (2000-5000€)
  - Pro (>5000€)

## 3. TRANSFORMACIONES EN SEGUIMIENTO

### Limpieza General
- Eliminadas columnas irrelevantes (`irrelevante_1`, `irrelevante_2`, `irrelevante_3`)
- Conversión de fechas con `pd.to_datetime()`
- Conversión de booleanos: conversion, cambio_curso, financiacion

### Procesamiento de Datos Numéricos
- Limpieza de columnas numéricas:
  - Manejo de valores 'nan', 'null', 'None'
  - Eliminación de espacios
  - Conversión con `pd.to_numeric()`
- Eliminación de valores negativos (reemplazados por NaN)

### Normalización de Categorías
- **estado_final**: Aplicado `strip()` y `title()`
- **metodo_pago**: Normalizado a Title Case
- **modalidad_final**: Normalizado a Title Case
- **comercial** y **closer**: Normalizados a Title Case

### Categorización
- **probabilidad_retorno**: Convertida a categorías
  - Bajo (0-30%)
  - Medio (30-70%)
  - Alto (70-100%)

### Deduplicación
- Eliminados duplicados por `seguimiento_id`

## 4. TRANSFORMACIONES EN PRIMER CONTACTO

### Limpieza de Datos
- Eliminadas columnas irrelevantes
- Limpieza exhaustiva de texto:
  - Eliminación de espacios extras
  - Reemplazo de valores como 'null', 'nan', 'n/a' por NaN
- Conversión de tipos:
  - Fechas a datetime
  - Hora a formato `HH:MM`
  - Datos numéricos

### Procesamiento Específico
- **presupuesto_confirmado**: 
  - Limpieza de valores no numéricos
  - Conversión a float o NaN
- **interes_otros_cursos**: 
  - Convertido a booleano
  - 'no', 'no aplica', 'ninguno' → `False`

### Imputación de Valores
- **Duración y nivel de interés**: Rellenados con la mediana
- **Campos de texto vacíos**: Rellenados con "No especificado"
- **preferencia_fecha_inicio**: Mapeada a fechas concretas
  - "lo antes posible" → fecha actual + 7 días
  - "en los próximos 3 meses" → fecha actual + 90 días
  - "en más de 3 meses" → fecha actual + 120 días
  - "aún no lo tiene claro" → fecha actual + 180 días
  - "no aplica" → NULL

## 5. TRANSFORMACIONES ADICIONALES PARA SQL

### Normalización de IDs
- **Mapeo de IDs de cursos**: 
  - IDs alfanuméricos convertidos a numéricos (1-7)
  - Creación de tabla de mapeo para mantener trazabilidad
- **Normalización de IDs principales**:
  - `lead_id`: Convertidos a enteros secuenciales
  - `contacto_id`: Convertidos a enteros secuenciales
  - `seguimiento_id`: Convertidos a enteros secuenciales

### Creación de Tabla de Tecnologías
- Extracción de 38 tecnologías únicas desde los cursos
- Creación de tabla normalizada con:
  - `tecnologia_id`: 1-38
  - `nombre`: Nombre de la tecnología

### Ajustes Finales para Base de Datos
- **Tipos de datos**: 
  - Conversión a tipos compatibles con SQL
  - Manejo de Int64 para valores nullable
- **Integridad referencial**:
  - Validación de claves foráneas entre tablas
  - Reemplazo por NULL de referencias no válidas
- **Columnas eliminadas**:
  - Columnas derivadas (Descuento_porcentaje, Horas_por_mes)
  - Categorías calculadas (grupo_edad, categoria_presupuesto)
  - Campos temporales de validación

### Archivos Generados
- `leads_final.csv`
- `cursos_final.csv`
- `seguimiento_final.csv`
- `primer_contacto_final.csv`
- `tecnologias_final.csv`

```python
df_leads ['presupuesto_estimado']
```

```python
if 'lead_score' in df_leads.columns:
    df_leads['lead_score'] = pd.to_numeric(df_leads['lead_score'], errors='coerce')
    df_leads['lead_score'] = df_leads['lead_score'].apply(lambda x: np.nan if pd.notna(x) and x > 100 else x)
    df_leads['lead_score'] = df_leads['lead_score'].astype('Int64')
```

```python
if 'codigo' in df_leads.columns:
    df_leads['codigo'] = df_leads['codigo'].astype(int)
```

```python
text_columns = ['nombre', 'apellidos', 'email', 'ciudad', 'provincia', 'nivel_estudios', 'situacion_laboral', 'objetivo_profesional','dispositivo_registro', 'horario_preferido', 'modalidad_preferida',
'comercial_asignado', 'fuente', 'campana']
```

```python
for col in text_columns:
    if col in df_leads.columns:
        df_leads[col] = df_leads[col].str.strip()
        if col in ['nombre', 'apellidos', 'ciudad', 'provincia', 'comercial_asignado']:
            df_leads[col] = df_leads[col].str.title()
```

```python
df_leads['email'] = df_leads['email'].str.lower()
```

```python
if 'lead_id' in df_leads.columns:
    duplicados_id = df_leads['lead_id'].duplicated().sum()
```

```python
if 'email' in df_leads.columns:
    duplicados_email = df_leads['email'].duplicated().sum()
    print(f"Duplicados por email: {duplicados_email}")
    if duplicados_email > 0:
        print("Emails duplicados:")
        print(df_leads[df_leads['email'].duplicated(keep=False)]['email'].value_counts())
```

```python
duplicados_totales = df_leads.duplicated().sum()
if duplicados_totales > 0:
    df_leads = df_leads.drop_duplicates()
```

```python
if 'edad' in df_leads.columns:
    df_leads['grupo_edad'] = pd.cut(df_leads['edad'], 
                             bins=[0, 25, 35, 45, 55, 100], 
                             labels=['18-25', '26-35', '36-45', '46-55', '56+'],
                             right=False)
```

```python
if 'presupuesto_estimado' in df_leads.columns:
    df_leads['categoria_presupuesto'] = pd.cut(df_leads['presupuesto_estimado'], 
                                        bins=[0, 3000, 5000, 7000, 100000], 
                                        labels=['Bajo', 'Medio', 'Alto', 'Pro'],
                                        right=False)
```

```python
if 'lead_score' in df_leads.columns:
    df_leads['calidad_lead'] = pd.cut(df_leads['lead_score'], 
                               bins=[0, 40, 60, 80, 100], 
                               labels=['Bajo', 'Medio', 'Alto', 'Excelente'],
                               right=False)
```

# Analizamos los datos del fichero 'Cursos'

```python
df_cursos = pd.read_csv('../data/Cursos.csv',
                          sep=';',
                          header=0,
                          index_col=False,
                          encoding='utf-8')
```

```python
df_cursos
```

```python
df_cursos = df_cursos.dropna(subset=['Curso ID'])
df_cursos = df_cursos.dropna(how='all')
```

```python
df_cursos
```

```python
text_columns = ['Nombre', 'Modalidad', 'Tecnologias', 'Nivel', 'Requisitos previos', 'Descripción', 'Certificación oficial']
for col in text_columns:
    if col in df_cursos.columns:
        df_cursos[col] = df_cursos[col].astype(str).str.strip().str.replace(r'\s+', ' ', regex=True)
```

```python
df_cursos.shape
```

```python
df_cursos.info()
```

```python
numeric_columns = ['Duracion meses', 'Horas totales', 'Precio base', 'Precio descuento', 
                  'Plazas disponibles', 'Rating', 'Empleabilidad']

for col in numeric_columns:
    if col in df_cursos.columns:
        df_cursos[col] = pd.to_numeric(df_cursos[col], errors='coerce')
```

```python
if 'Fecha inicio proxima' in df_cursos.columns:
        df_cursos['Fecha inicio proxima'] = pd.to_datetime(df_cursos['Fecha inicio proxima'], errors='coerce')
        df_cursos['Fecha inicio proxima'] = df_cursos['Fecha inicio proxima'].dt.strftime('%Y-%m-%d')
```

```python
if 'Precio base' in df_cursos.columns and 'Precio descuento' in df_cursos.columns:
    df_cursos['Descuento_porcentaje'] = ((df_cursos['Precio base'] - df_cursos['Precio descuento']) / df_cursos['Precio base'] * 100).round(2)
```

```python
if 'Horas totales' in df_cursos.columns and 'Duracion meses' in df_cursos.columns:
    df_cursos['Horas_por_mes'] = (df_cursos['Horas totales'] / df_cursos['Duracion meses']).round(2)
```

```python
if 'Precio base' in df_cursos.columns:
    df_cursos['Categoria_precio'] = pd.cut(df_cursos['Precio base'], 
                                         bins=[0, 2000, 5000, float('inf')], 
                                         labels=['Económico', 'Medio', 'Pro'], 
                                         include_lowest=True)
```

```python
df_cursos = df_cursos.reset_index(drop=True)
```
